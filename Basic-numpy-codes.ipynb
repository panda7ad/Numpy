# Stacking and Splitting


### a) Horizontal Stacking(hstack)
##### combines array side by side

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(np.hstack((a, b)))

```

### 

### b) Vertical Stacking(vstack)
#### combines arrays vertically

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(np.vstack((a, b)))
```

### c) Column stack
#### makes arrays in columns

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(np.column_stack((a, b)))
```

# Splitting Array
#### Splits one array into multiple parts

```python
arr = np.array([1, 2, 3, 4, 5, 6])

result = np.split(arr, 3)

print(result)
```
